```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos originales
tam = ["Mil", "Diez Mil", "Cien Mil", "Un Millón"]
seleccion = [1526, 128949, 16237079, 1254679998]
insercion = [726, 64341, 6377704, 639481776]
mezcla = [194, 1472, 15405, 181881]
monticulos = [206, 2266, 28640, 335299]
rapido = [120, 2326, 92905, 12451941]
residuos = [1075, 757, 6705, 68818]
```

```python
# Aplicar log base 10
datos = {
    'Selección': np.log10(seleccion),
    'Inserción': np.log10(insercion),
    'Mezcla': np.log10(mezcla),
    'Montículos': np.log10(monticulos),
    'Rápido': np.log10(rapido),
    'Residuos': np.log10(residuos)
}
```

```python
valores_originales = {
    'Selección': seleccion,
    'Inserción': insercion,
    'Mezcla': mezcla,
    'Montículos': monticulos,
    'Rápido': rapido,
    'Residuos': residuos
}
```

```python

# Crear subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 10))
fig.subplots_adjust(hspace=0.5)

for ax, (nombre, tiempos) in zip(axs.ravel(), datos.items()): 
    ax.plot(tam, tiempos, marker='o')
    ax.set_title(nombre)
    ax.set_xlabel("Tamaño del Arreglo")
    ax.set_ylabel("log(Tiempo)")

    # Líneas horizontales grises
    for valor in tiempos:
        ax.axhline(y=valor, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
    
    # Etiquetas en cada punto
    for i, valor in enumerate(tiempos):
        # Etiqueta superior con tiempo original + µs
        ax.annotate(f"{valores_originales[nombre][i]} µs", 
                    (tam[i], valor), 
                    textcoords="offset points", 
                    xytext=(0, 5), 
                    ha='center', fontsize=8, color='black')
        
        # Etiqueta inferior con valor logarítmico
        ax.annotate(f"log: {valor:.2f}", 
                    (tam[i], valor), 
                    textcoords="offset points", 
                    xytext=(0, -10), 
                    ha='center', fontsize=8, color='gray')

# Resultados
plt.suptitle("Comparación logarítmica de algoritmos de ordenamiento")
plt.show()
```
